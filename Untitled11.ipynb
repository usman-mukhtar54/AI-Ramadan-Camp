```python
num1 = int(input("enter first value: "))
operator = input("enter operator (+, -, *, /): ")
num2 = int(input("enter second value: "))
    
if operator == '+':
     print(f"{num1} + {num2} = {num1 + num2}")
elif operator == '-':
     print(f"{num1} - {num2} = {num1 - num2}")
elif operator == '*':
     print(f"{num1} * {num2} = {num1 * num2}")
elif operator == '/':
    if num2 != 0:
         print(f"{num1} / {num2} = {num1 / num2}")
    else:
        print("error: division by zero!")
else:
    print("invalid operator!")

```

```python
my_list = [1, 2, 3, 2, 4, 5, 1]
duplicates = []
unique = []
for x in my_list:
    if x in unique:
        duplicates.append(x)
    else:
        unique.append(x)
print(duplicates)
```
